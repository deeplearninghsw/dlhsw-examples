```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from utils.plotting_KMeans import plot

from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('data/ex.csv')

dataset = df.copy()
```

```python
kmeans = KMeans(n_clusters=3, max_iter=3_000)
kmeans.fit(dataset.values)
y_pred = kmeans.predict(dataset.values)

plot(dataset.values, dataset['y'].values, y_pred, kmeans.cluster_centers_, kmeans)

score = kmeans.score(dataset.values, dataset['y'].values)

print(f"Score: {score}")
```
